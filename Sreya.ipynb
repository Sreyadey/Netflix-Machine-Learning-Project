```python
import pandas as pd # used for linear algebra oprations
import numpy as np #used for data preparation
import plotly.express as px # used for data visualization
import textblob as textblob #used for sentiment analysis
```

```python
df=pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

# NOW TO DETERMINE THE RATING OF CONTENT IN DATASHEET.


```python
x=df.groupby(['rating']).size().reset_index(name='counts')
print(x)
```

LET'S UNDERSTAND BY **PIECHART**


```python
pieChart=px.pie(x,values='counts',names='rating',title='Distribution of Content Ratings')
pieChart.show()
```

AS WE SEE MOST OF THE DIRECTOR NAME ARE MENTION AS NAN ,SO WE HAVE TO FILL DIRECTOR UNSPECIFIED.

```python
df['director'].fillna(' Director Unspecified',inplace=True)
df.head()
```

NOW WE HAVE TO TAKE DIRECTOR NAME IN ANOTHER ROW AND COLUMNS.


```python
directors_list =pd.DataFrame()
print(directors_list)
```

NOW  WE HAVE TO FILL THE DATA IN IT


```python
directors_list=df['director'].str.split(',',expand=True).stack()
print(directors_list)
```

## *NOW TO FRAME DIRECTOR DATA IN ROW AND COLUMNS*


```python
directors_list=directors_list.to_frame()
directors_list.columns=['director']
directors_list.head()
```

## **Let italicized text's count the all the directors in this content**


```python
directors=directors_list.groupby(['director']).size().reset_index(name='counts')
print(directors)
```

*NOW WE HAVE TO REMOVE THE DIRECTORS WHICH ARE UNSPECIFIED*


```python
directors = directors[directors.director != ' Director Unspecified']
print(directors)
```

WE HAVE TO ARRANGE IN ASCENDING MANNER


```python
directors=directors.sort_values(by=['counts'])
print(directors)
```

TO FIND IN DECENDING MANNER

```python
directors=directors.sort_values(by=['counts'],ascending=False)
print(directors)
```

TO TAKE TOP FIVE DIRECTORS


```python
top5directors=directors[:5]
print(top5directors)
```

NOW WE HAVE TO CREATE HORIZONTAL BAR CHART TO KNOW HOW MUCH CONTENT THEY CREATED


```python
top5directors= top5directors.sort_values(by=['counts'])
barChart=px.bar(top5directors,x='counts',y='director',title='Top 5 Directors with Most Content',orientation='h')
barChart.show()
```

NOW ANALYSIS THE TOP 5 ACTORS IN NETFLIX


```python
df['cast']=df['cast'].fillna('Actor Unspecified')
cast_df=pd.DataFrame()
cast_df=df['cast'].str.split(',',expand=True).stack()
cast_df=cast_df.to_frame()
cast_df.columns=['Actors']
actors=cast_df.groupby(['Actors']).size().reset_index(name='counts')
actors=actors[actors.Actors!='Actor Unspecified']
actors=actors.sort_values(by=['counts'],ascending=False)
top5actors=actors.head()
top5actors=top5actors.sort_values(by=['counts'])
barChart2=px.bar(top5actors,x='counts',y='Actors',title='Top 5 Actors with Most Content',orientation='h')
barChart2.show()
```

**SENTIMENT ANALYSIS OF NETFLIX CONTENT**


```python
df2=df[['release_year','description']]
df2=df2.rename(columns={'release_year':'Release Year', 'description':'Description'})
for index,row in df2.iterrows():
  d=row['Description']
  testimonial=textblob.TextBlob(d)
  p=testimonial.sentiment.polarity
  if p==0:
    sent='Neutral'
  elif p>0:
    sent='Positive'
  else:
    sent='Negative'
  df2.loc[[index,2], 'sentiment']=sent

df2=df2.groupby(['Release Year','sentiment']).size().reset_index(name='counts')
df2=df2[df2['Release Year']>2005]
barGraph=px.bar(df2,x='Release Year',y='counts',color='sentiment',title='Sentiment Analysis of Netflix Content')
barGraph.show()
```
